```python
import roboticstoolbox as rtb
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import spatialmath.base as base
```

```python
ur5 = rtb.models.DH.UR5()
```

```python
#Muestra del proceso iterativo del movimiento del robot en el espacio
from roboticstoolbox import mtraj, trapezoidal
mtrajU = mtraj(trapezoidal, ur5.qz, ur5.qr, 10)
timem = mtrajU.t
confm = mtrajU.q 

for i in range(len(timem)):
    print('Time ', timem[i])
    print('Conf: ', confm[i])
```

```python

Tz = ur5.fkine(ur5.qz)
Tr= ur5.fkine(ur5.qr)
ctraj = rtb.ctraj(Tz, Tr, 100)
```

```python
#Muestra del proceso iterativo de las juntas en el espacio
jtraj = rtb.jtraj(ur5.qz, ur5.qr, t= 100)
time = jtraj.t
confs = jtraj.q

for i in range(len(time)):
    print('Time ', time[i])
    print('Conf: ', confs[i])
```
